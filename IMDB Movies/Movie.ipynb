-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.

3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here

dropDuplicateColumns['genre_1'] = dropDuplicateColumns.genres.str.split('|', expand=True)[0]
dropDuplicateColumns['genre_2'] = dropDuplicateColumns.genres.str.split('|', expand=True)[1]

## check that genre 2 has nan values and replace it with genre_1 value using inplace.
print('NaN values in genre_2 column before replacing: ', dropDuplicateColumns['genre_2'].isnull().sum())
dropDuplicateColumns.genre_2.fillna(dropDuplicateColumns.genres,inplace=True)
print('NaN values in genre_2 column after replacing: ', dropDuplicateColumns['genre_2'].isnull().sum())
```

```python
movies_by_segment = dropDuplicateColumns.groupby(['genre_1','genre_2'])
```

```python
PopGenre = dropDuplicateColumns.groupby(['genre_1','genre_2'])['gross'].agg(['mean'])
PopGenre = PopGenre.sort_values('mean', ascending=False)
print('The top 5 genre combinations are:')
PopGenre.head(5)
```

```python
# Write the code for calculating decade here

## To obtain the correct decade for a movie we can perform floor value division of the title year by 10 
## & multiply it by 10 to obtain the correct decade. 
dropDuplicateColumns['decade'] = (dropDuplicateColumns.title_year//10)*10
dropDuplicateColumns.head()
```

```python
# Write your code for creating the data frame df_by_decade here 
df_by_decade = dropDuplicateColumns.groupby(['decade'], sort=True).sum().reset_index()
df_by_decade
```

```python
# Write your code for plotting number of voted users vs decade

## before we plot the bar graph, we must take cognizance of 
## the very high difference between the highest and lowest values in num_voted_users

## to prevent this, we will adjust the column by dividing it by 1000.
df_by_decade['num_voted_users'] = (df_by_decade['num_voted_users'].astype(float)/1000)

barPlot = df_by_decade.plot.bar(x='decade', y='num_voted_users', rot=0, figsize=(15,15))
barPlot.set_xlabel('Decade')
barPlot.set_ylabel('Number of voters')
```

### Observations:
Given the bar plot above, we can safely say the following:
1. The number of voters grew to almost double every decade till the 2010's decade
2. The number of voters grew from 1920's to 1930's but fell drastically in 1940's
3. The number of voters has seen a steady rise of almost doubling every decade.
4. The decade with the highest vote count is 2000s, with 165976645 users
